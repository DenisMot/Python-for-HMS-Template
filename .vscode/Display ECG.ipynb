ECG signal

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

Import an EGC data

```python
# load the data
ECG_data = np.loadtxt(r"C:\Users\theol\OneDrive\Documents\GitHub\IEAP\Python-R-Git\Series01-Python\data\ECGu.txt")

# look at the data
print(ECG_data)
print(ECG_data.shape)
```

Convert raw data

```python
#Conversion to mV
ECG_mV = ECG_data*1024

#Conversion to seconds
ts = 1000
t = np.arange(ECG_mV.shape[0]) / ts

```

Diplay ECG signal

```python
plt.figure(figsize=(7,5))   #Choose a good size to read quickly
plt.grid(True)              #Place a grid
plt.plot(t,ECG_mV)          #Place x and y 
plt.title("ECG Signal")     #Name title, x and y 
plt.xlabel("Time (s)")      
plt.ylabel("ECG (mV)")
```

Create a realistic ECG plot

```python
plt.figure(figsize=(12,5))   
plt.grid(True,which='major', color='red',linestyle='-', linewidth= 0.8) #
plt.minorticks_on()
plt.grid(True,which='minor', color='red',linestyle='--', linewidth= 0.3)
plt.plot(t,ECG_mV[:,0])        
plt.title("ECG Sigal I")     
plt.xlabel("Time (s)")      
plt.ylabel("ECG (mV)")

plt.figure(figsize=(12,5))  
plt.grid(True,which='major', color='red',linestyle='-', linewidth= 0.8)
plt.minorticks_on()
plt.grid(True,which='minor', color='red',linestyle='--', linewidth= 0.3)              
plt.plot(t,ECG_mV[:,1])         
plt.title("ECG Signal II")      
plt.xlabel("Time (s)")
plt.ylabel("ECG (mV)")

plt.figure(figsize=(12,5))  
plt.grid(True,which='major', color='red',linestyle='-', linewidth= 0.8)
plt.minorticks_on()
plt.grid(True,which='minor', color='red',linestyle='--', linewidth= 0.3)              
plt.plot(t,ECG_mV[:,2])          
plt.title("ECG Signal III")     
plt.xlabel("Time (s)")
plt.ylabel("ECG (mV)")
```
